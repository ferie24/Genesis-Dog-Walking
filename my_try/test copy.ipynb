```python
%load_ext autoreload
%autoreload 2

import logging
import genesis as gs

gs.init(logging_level=logging.WARNING, backend=gs.gpu)
```

```python
from environment.make_environment import Go2WalkingEnv
from runner import Runner
from policy.ppo import PPO
from policy.rewards import Rewards

reward_fn = Rewards()
num_envs = 2000

# Create environment
env = Go2WalkingEnv(
    num_envs=num_envs,
    device="cuda",
    show_viewer=False,
    use_terrain=False,  # Set to True for complex terrain
    episode_length_s=20.0,
    reward_fn=reward_fn
)
```

```python

# Set walking command (1.0 m/s forward)
env.set_commands(lin_vel_x=1.0, lin_vel_y=0.0, ang_vel_yaw=0.0)

policy = PPO(
    num_envs=env.num_envs,
    num_obs=env.num_obs,
    num_actions=env.num_actions,
    device=env.device,
)

runner = Runner(env, policy)

# Reset environment
obs = env.reset()

```
